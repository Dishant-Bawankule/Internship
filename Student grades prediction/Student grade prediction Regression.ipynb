As we can see there are many null values in our dataset first we need to handel that null values.

```python
# Dropping rows with any missing values
df = df.dropna()

# Dropping columns with more than 50% missing values
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)
```

```python
df.shape
```

```python
df.isnull().sum()
```

We droped the null all null value from the data which are having more null values.

```python
# checking the value count for each column
for i in df.columns:
    print(df[i].value_counts())
    print('\n')
```

Here in each column there are several categories in object datatype so we need to convert those datatype in number data for model building.

we will assign some numbers to to the several category in our dataset.

```python
# Define the values that need to be dropped
values_to_drop = ['W', 'WU']

# Drop rows containing 'W' or 'WU' in any column
df = df[~df.isin(values_to_drop).any(axis=1)]

# Display the first few rows of the cleaned dataset
df.shape
```

Grade A+:- 4.3, A:- 4.0, A-:- 3.7, B+:- 3.3, B:-3.0, B-:-2.7, C+:- 2.3, C:- 2.0, C-:- 1.7, D+:- 1.3, D:- 1.0, D-:- 0.7, F:-0.0

```python
# Define a mapping for letter grades to numerical values
grade_mapping = {
    'A+': 4.3, 'A': 4.0, 'A-': 3.7, 'B+': 3.3, 'B': 3.0, 'B-': 2.7,
    'C+': 2.3, 'C': 2.0, 'C-': 1.7, 'D+': 1.3, 'D': 1.0, 'D-': 0.7, 'F': 0.0,
    'A+': 4.0, 'C+R': 2.3 }

# Apply the mapping to the relevant columns
for col in df.columns[1:-1]:  # Exclude 'Seat No.' and 'CGPA'
    df[col] = df[col].map(grade_mapping)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
maxR2_score=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(X_scaler,y,test_size = 0.25, random_state=i)
    lin_reg=LinearRegression()
    lin_reg.fit(x_train,y_train)
    y_pred=lin_reg.predict(x_test)
    R2=r2_score(y_test,y_pred)
    if R2>maxR2_score:
        maxR2_score=R2
        maxRS=i
print('Best accuracy is', maxR2_score ,'on Random_state', maxRS)
```

# Linear Regression

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaler,y,test_size = 0.25, random_state=148)
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
lin_reg.score(x_train,y_train)
y_pred=lin_reg.predict(x_test)
print('Predicted result :\n', y_pred)
print('\n')
print('actual result\n',y_test)
```

# Evaluation Metrics of linear regression

```python
print('\033[1m'+' Error :'+'\033[0m')
print('Mean absolute error :', mean_absolute_error(y_test,y_pred))
print('Mean squared error :', mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
print('\n')
from sklearn.metrics import r2_score
print('\033[1m'+' R2 Score :'+'\033[0m')
print(r2_score(y_test,y_pred))
```

```python
rf = RandomForestRegressor(n_estimators = 100 )
svr=SVR()
dtc = DecisionTreeRegressor()
ad=AdaBoostRegressor()

model = [rf,svr,dtc,ad]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    y_pred = m.predict(x_test)
                                            
    print('\033[1m'+' Error of ', m, ':' +'\033[0m')
    print('Mean absolute error :', mean_absolute_error(y_test,y_pred))
    print('Mean squared error :', mean_squared_error(y_test,y_pred))
    print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,y_pred)))
    print('\n')

    print('\033[1m'+' R2 Score :'+'\033[0m')
    print(r2_score(y_test,y_pred)) 
    print('\n')
```

As we can see linear regression is giving better result.
